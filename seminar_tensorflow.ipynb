```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasets/cars_sample.csv.zip')
```

```python
df.columns.tolist()
```

```python
_ = df.pop('model')
y = df.pop('price')
```

```python
categs = ['region','year','manufacturer','condition','cylinders','fuel','transmission','drive','size','type','paint_color']
```

```python
df_sparse = pd.get_dummies(df, columns=categs)
df_sparse.shape
```

```python
df_sparse.iloc[:10, :15]
```

```python
X = df_sparse.values
```

```python
X.shape
```

```python
X[:,0] = (X[:,0] - X[:,0].min()) / (X[:,0].max() - X[:,0].min()) #study: other types of scaling
```

```python
X_train = X[:-100]
y_train = y[:-100]
X_test = X[-100:]
y_test = y[-100:]
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
#input
tf_inp = tf.keras.layers.Input(shape=X.shape[1], name='inp')
#corpus
tf_x = tf_inp
tf_x = tf.keras.layers.Dense(units=256, name='dense1', activation='relu')(tf_x)
tf_x = tf.keras.layers.Dense(units=128, name='dense2', activation='relu')(tf_x)
tf_x = tf.keras.layers.Dense(units=64, name='dense3', activation='relu')(tf_x)
#output
tf_out = tf.keras.layers.Dense(units=1, name='readout')(tf_x)

model = tf.keras.models.Model(inputs=tf_inp, outputs=tf_out)
model.compile(loss='mae', optimizer='sgd')
model.summary()

model.fit(x=X_train, y=y_train, epochs=200, batch_size=32, validation_data=(X_test, y_test))
```

```python
y_preds = model.predict(X_test).ravel()
```

```python
mae = np.abs(y_preds - y_test).mean()
```
